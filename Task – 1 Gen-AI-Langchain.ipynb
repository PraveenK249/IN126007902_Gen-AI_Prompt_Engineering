```python
from langchain_core.prompts import ChatPromptTemplate

# Define role-based system messages
role_messages = {
    "teacher": "You are an expert teacher who explains concepts clearly and patiently.",
    "interviewer": "You are a professional interviewer asking insightful questions.",
    "motivator": "You are an inspiring motivator who encourages and energizes people."
}

def create_chat_prompt(role, topic):
    """
    Create a chat prompt with system and user messages
    """
    chat_template = ChatPromptTemplate.from_messages([
        ("system", role_messages[role]),
        ("human", "Tell me about {topic}")
    ])
    
    return chat_template.format_messages(topic=topic)

# Test cases
test_scenarios = [
    {"role": "teacher", "topic": "Neural Networks"},
    {"role": "interviewer", "topic": "Neural Networks"},
    {"role": "motivator", "topic": "Neural Networks"}
]

print("=" * 60)
print("TASK 4: ChatPromptTemplate System")
print("=" * 60)

for scenario in test_scenarios:
    role = scenario["role"]
    topic = scenario["topic"]
    
    print(f"\nRole: {role.upper()}")
    print(f"Topic: {topic}")
    print("-" * 60)
    
    messages = create_chat_prompt(role, topic)
    
    for msg in messages:
        print(f"{msg.type.upper()}: {msg.content}")
    
print("=" * 60)
```

```python
def validate_inputs(audience, tone, use_defaults=True):
    """
    Validate audience and tone inputs
    
    Args:
        audience: Target audience level
        tone: Communication tone
        use_defaults: If True, assign defaults; if False, raise error
    
    Returns:
        tuple: (validated_audience, validated_tone)
    """
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]
    
    # Validation flags
    audience_valid = audience.lower() in valid_audiences
    tone_valid = tone.lower() in valid_tones
    
    # Handle invalid inputs
    if not audience_valid or not tone_valid:
        if use_defaults:
            # Assign default values
            if not audience_valid:
                print(f"⚠️  Invalid audience '{audience}'. Using default: 'beginner'")
                audience = "beginner"
            if not tone_valid:
                print(f"⚠️  Invalid tone '{tone}'. Using default: 'formal'")
                tone = "formal"
        else:
            # Raise error
            if not audience_valid:
                raise ValueError(f"Invalid audience: '{audience}'. Must be one of {valid_audiences}")
            if not tone_valid:
                raise ValueError(f"Invalid tone: '{tone}'. Must be one of {valid_tones}")
    
    return audience.lower(), tone.lower()

# Test validation
print("=" * 60)
print("TASK 5: Input Validation Layer")
print("=" * 60)

# Valid inputs
print("\n1. Valid Inputs:")
audience, tone = validate_inputs("beginner", "formal")
print(f"   ✅ Audience: {audience}, Tone: {tone}")

# Invalid inputs with defaults
print("\n2. Invalid Inputs (with defaults):")
audience, tone = validate_inputs("advanced", "aggressive", use_defaults=True)
print(f"   ✅ Corrected - Audience: {audience}, Tone: {tone}")

# Invalid inputs with error
print("\n3. Invalid Inputs (with error):")
try:
    audience, tone = validate_inputs("phd", "angry", use_defaults=False)
except ValueError as e:
    print(f"   ❌ Error: {e}")

print("=" * 60)
```

```python
from langchain_core.prompts import PromptTemplate

def generate_prompt(topic, audience, tone, style):
    """
    Generate dynamic prompts based on inputs
    
    Args:
        topic: The subject to explain
        audience: Target audience level
        tone: Communication tone
        style: Presentation style
    
    Returns:
        str: Generated prompt
    """
    # Validate inputs (suppress output for cleaner display)
    try:
        audience, tone = validate_inputs(audience, tone, use_defaults=True)
    except:
        audience, tone = "beginner", "formal"
    
    # Define style templates
    style_templates = {
        "teaching": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} tone, using a step-by-step teaching approach"
        ),
        "interview": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Prepare interview questions about {topic} suitable for {audience} level in a {tone} tone"
        ),
        "storytelling": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} storytelling style"
        )
    }
    
    # Get appropriate template
    if style.lower() not in style_templates:
        print(f"⚠️  Invalid style '{style}'. Using default: 'teaching'")
        style = "teaching"
    
    template = style_templates[style.lower()]
    
    # Generate prompt
    prompt = template.format(topic=topic, audience=audience, tone=tone)
    
    return prompt

# Test the generator
print("=" * 60)
print("TASK 6: Prompt Generator App")
print("=" * 60)

test_cases = [
    {
        "topic": "Neural Networks",
        "audience": "beginner",
        "tone": "fun",
        "style": "storytelling"
    },
    {
        "topic": "Data Structures",
        "audience": "intermediate",
        "tone": "formal",
        "style": "teaching"
    },
    {
        "topic": "Cloud Computing",
        "audience": "expert",
        "tone": "casual",
        "style": "interview"
    }
]

for i, test in enumerate(test_cases, 1):
    print(f"\nExample {i}:")
    print(f"  Topic: {test['topic']}")
    print(f"  Audience: {test['audience']}")
    print(f"  Tone: {test['tone']}")
    print(f"  Style: {test['style']}")
    
    prompt = generate_prompt(**test)
    print(f"\n  Generated Prompt:")
    print(f"  {prompt}")
    print("-" * 60)

print("=" * 60)
```

```python
from langchain_core.prompts import PromptTemplate

def complete_pipeline(topic, audience, tone, style):
    """
    Complete pipeline: Input → Validation → Template → Output
    """
    print("=" * 60)
    print("COMPLETE PIPELINE FLOW")
    print("=" * 60)
    
    # Step 1: User Input
    print("\n📥 STEP 1: User Input")
    print(f"   Topic: {topic}")
    print(f"   Audience: {audience}")
    print(f"   Tone: {tone}")
    print(f"   Style: {style}")
    
    # Step 2: Validation
    print("\n✅ STEP 2: Validation")
    validated_audience, validated_tone = validate_inputs(audience, tone, use_defaults=True)
    print(f"   Validated Audience: {validated_audience}")
    print(f"   Validated Tone: {validated_tone}")
    
    # Step 3: Prompt Template Selection
    print("\n📝 STEP 3: Prompt Template Selection")
    print(f"   Selected Style: {style}")
    
    # Step 4: Dynamic Prompt Generation
    print("\n🔄 STEP 4: Dynamic Prompt Generation")
    prompt = generate_prompt(topic, validated_audience, validated_tone, style)
    
    # Step 5: Output
    print("\n📤 STEP 5: Final Output")
    print(f"   Generated Prompt:")
    print(f"   '{prompt}'")
    
    print("\n" + "=" * 60)
    print("✅ Pipeline Completed Successfully!")
    print("=" * 60)
    
    return prompt

# Run complete pipeline
result = complete_pipeline(
    topic="Neural Networks",
    audience="beginner",
    tone="fun",
    style="storytelling"
)
```
